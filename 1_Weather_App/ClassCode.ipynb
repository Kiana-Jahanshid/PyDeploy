```python
import http.client

conn = http.client.HTTPSConnection("goweather.herokuapp.com")
payload = ''
headers = {}
conn.request("GET", "/weather/barare", payload, headers)
res = conn.getresponse()
data = res.read()
print(data.decode("utf-8"))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
response = requests.get("https://picsum.photos/300/300")
file_data = np.frombuffer(response.content , dtype="int8")
image= cv2.imdecode(file_data , cv2.IMREAD_UNCHANGED)
plt.imshow(image)
```
